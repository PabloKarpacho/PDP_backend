```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# --- 1. Создадим исходные данные ---
x = np.linspace(-5, 5, 100)
y_true = np.sin(x) + 0.1 * np.random.randn(len(x))  # функция с шумом

# --- 2. Выбираем центры RBF ---
num_centers = 10
centers = np.linspace(-5, 5, num_centers)

# --- 3. Вычисляем RBF активации (используем Гаусс) ---
def rbf(x, c, sigma):
    return np.exp(-((x - c)**2) / (2 * sigma**2))

sigma = 1.0  # ширина гаусса
Phi = np.array([rbf(x, c, sigma) for c in centers]).T  # матрица признаков

# --- 4. Обучаем линейные веса выходного слоя ---
# y = Phi @ w => находим w методом наименьших квадратов
w = np.linalg.pinv(Phi) @ y_true

# --- 5. Получаем аппроксимацию ---
y_pred = Phi @ w

# --- 6. Визуализация аппроксимации ---
plt.figure(figsize=(12, 6))
plt.scatter(x, y_true, color='blue', label='Исходные данные', alpha=0.5)
plt.plot(x, y_pred, color='red', linewidth=2, label='Аппроксимация RBF')
plt.scatter(centers, np.zeros_like(centers), color='green', marker='x', s=100, label='Центры RBF')

# --- 7. Визуализация поведения каждой RBF вокруг центра ---
for c in centers:
    plt.plot(x, rbf(x, c, sigma), linestyle='--', alpha=0.5, label=f'RBF центр {c:.1f}')

plt.title('Аппроксимация функции с использованием RBF сети и поведение RBF нейронов')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Генерация данных ---
np.random.seed(42)
N = 50  # точек на класс
class_A = np.random.randn(N, 2) + np.array([2, 2])
class_B = np.random.randn(N, 2) + np.array([-2, -2])
X = np.vstack([class_A, class_B])
y = np.array([0]*N + [1]*N)  # метки классов

# --- Инициализация прототипов ---
num_prototypes_per_class = 2
prototypes = np.array([class_A[np.random.choice(N)], class_A[np.random.choice(N)],
                       class_B[np.random.choice(N)], class_B[np.random.choice(N)]])
proto_labels = np.array([0, 0, 1, 1])

# --- Параметры обучения ---
alpha = 0.1
epochs = 10

# --- Функция обучения LVQ1 ---
for epoch in range(epochs):
    for i, x_i in enumerate(X):
        # находим ближайший прототип
        dists = np.linalg.norm(prototypes - x_i, axis=1)
        winner_idx = np.argmin(dists)
        # корректируем прототип
        if proto_labels[winner_idx] == y[i]:
            prototypes[winner_idx] += alpha * (x_i - prototypes[winner_idx])
        else:
            prototypes[winner_idx] -= alpha * (x_i - prototypes[winner_idx])

# --- Визуализация ---
plt.figure(figsize=(8, 6))
plt.scatter(class_A[:,0], class_A[:,1], color='red', label='Класс A')
plt.scatter(class_B[:,0], class_B[:,1], color='blue', label='Класс B')
plt.scatter(prototypes[:2,0], prototypes[:2,1], color='darkred', marker='x', s=100, label='Прототипы A')
plt.scatter(prototypes[2:,0], prototypes[2:,1], color='darkblue', marker='x', s=100, label='Прототипы B')
plt.title('LVQ1: обучение прототипов')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Простейшие данные ---
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

# --- Параметры простейшей модели ---
w = 0.5  # вес
b = 0.0  # смещение
eta = 0.001  # скорость обучения

# --- Функция предсказания ---
def predict(x, w, b):
    return w * x + b

# --- Функция потерь (MSE) ---
def loss(y_pred, y_true):
    return np.mean((y_pred - y_true)**2)

# --- 1. Batch Training ---
w_batch = w
b_batch = b
loss_batch_list = []
for epoch in range(5):
    y_pred = predict(x, w_batch, b_batch)
    grad_w = np.mean(2 * (y_pred - y) * x)  # средний градиент по всей выборке
    grad_b = np.mean(2 * (y_pred - y))
    w_batch -= eta * grad_w
    b_batch -= eta * grad_b
    current_loss = loss(y_pred, y)
    loss_batch_list.append(current_loss)
    print(f"Epoch {epoch+1} Batch Loss: {current_loss:.4f}")

# --- 2. Online Training ---
w_online = w
b_online = b
loss_online_list = []
for epoch in range(5):
    for xi, yi in zip(x, y):
        y_pred_i = predict(xi, w_online, b_online)
        grad_w = 2 * (y_pred_i - yi) * xi  # градиент для одного примера
        grad_b = 2 * (y_pred_i - yi)
        w_online -= eta * grad_w
        b_online -= eta * grad_b
    y_pred_all = predict(x, w_online, b_online)
    current_loss = loss(y_pred_all, y)
    loss_online_list.append(current_loss)
    print(f"Epoch {epoch+1} Online Loss: {current_loss:.4f}")

# --- Визуализация ---
plt.figure(figsize=(12, 5))

# График аппроксимации
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Истинные значения')
plt.plot(x, predict(x, w_batch, b_batch), color='red', label='Batch Prediction')
plt.plot(x, predict(x, w_online, b_online), color='green', linestyle='--', label='Online Prediction')
plt.title('Сравнение предсказаний')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
```
